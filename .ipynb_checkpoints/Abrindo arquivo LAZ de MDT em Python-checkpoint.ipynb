# Abrindo arquivo LAZ de MDT em Python

Vamos utilizar o Python para abrir o arquivo LAZ baixado do GeoSampa

```python
import laspy
import scipy
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
from sklearn import metrics
from sklearn import preprocessing
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import path
```

```python
!pdal translate ./mdts/MDT_3316-223_1000_v4.laz ./mdts/MDT_3316-223_1000_v4.las
```

```python
inFile = laspy.file.File("./mdts/MDT_3316-223_1000_v4.las")
inFile
```

```python
# Grab a numpy dataset of our clustering dimensions:
dataset = np.vstack([inFile.x, inFile.y, inFile.z]).transpose()
dataset.shape
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot3D(dataset[:, 0], dataset[:, 1], dataset[:, 2], 'gray')
```
